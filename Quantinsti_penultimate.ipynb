```python
import pandas as pd
import pandas_datareader.data as web
import numpy as np
import yfinance as yf
import talib as ta
import matplotlib.pyplot as plt

nvda = yf.Ticker("NVDA")

pd.options.display.max_rows = 500
pd.options.display.max_columns = 100

from dateutil import relativedelta
from datetime import datetime
import calendar

import warnings

warnings.filterwarnings('ignore')

# importing random forest classifier from assemble module
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, make_scorer, mean_absolute_error

from skopt import BayesSearchCV  # module for Bayesian optimisation on the scikit learn backend


import xgboost as xgb
from xgboost.sklearn import XGBClassifier

import lightgbm as lgb
```

```python
#fetching nvidia OHLCV data
df_nvda = nvda.history('max').loc['2019-01-01':'2024-12-31', :][[
    'Open', 'High', 'Low', 'Close', 'Volume'
]]
```

```python
# define the periods to calculate returns
intervals = [1, 5, 12, 22, 44, 88]

#Loop to calculate returns for time in intervals
for t in intervals:
    df_nvda.loc[:, 'ret_{}d'.format(t)] = df_nvda.Close.pct_change(t)
```

```python
#using ta-lib python module for calculation of different indicators

######################## volatility indicators

# Average True Range: volatility indicator that measures the average price movement range over a given period
df_nvda.loc[:, 'Norm_ATR'] = ta.NATR(df_nvda.High,
                                     df_nvda.Low,
                                     df_nvda.Close,
                                     timeperiod=14)

# Bolinger bands: volatility-based technical indicator
upper, middle, lower = ta.BBANDS(df_nvda.Close, timeperiod=20, nbdevup=2, nbdevdn=2)
df_nvda.loc[:, 'Upper_BB'] = upper
df_nvda.loc[:, 'Middle_BB'] = middle
df_nvda.loc[:, 'Lower_BB'] = lower

######################## momentum indicator
#Relative Strength Index (RSI) is a momentum oscillator
df_nvda.loc[:, 'RSI'] = ta.RSI(df_nvda.Close, timeperiod=14)

# macd signal: momentum indicator used to identify trend direction, strength, and potential reversals.
macd, macdsignal, macdhist = ta.MACD(df_nvda.Close,
                                     fastperiod=12,
                                     slowperiod=26,
                                     signalperiod=9)
df_nvda.loc[:, 'macd'] = macd
df_nvda.loc[:, 'macdsignal'] = macdsignal

######################## trend indicator
#ADX (Average Directional Index) is used to measure trend strength.
df_nvda.loc[:, 'ADX'] = ta.ADX(df_nvda.High,
                               df_nvda.Low,
                               df_nvda.Close,
                               timeperiod=14)
```

```python
#Creating Target Variable for ML model
#next day return brought backwards to create target variable for next day's returns
df_nvda.loc[:, 'return'] = df_nvda.Close.pct_change(1)
df_nvda.loc[:, 'target'] = np.where(df_nvda['return'].shift(-1) > 0, 1, 0)

df_nvda.loc[:, 'Year_month'] = (
    pd.Series(df_nvda.index).dt.strftime("%Y") + str('_') +
    pd.Series(df_nvda.index).dt.strftime("%m")).to_list()
```

```python
#Descriptive plot to visualise how often in each month the stock has moved up/down

plt_table = df_nvda.groupby(['Year_month'])['return'].apply(
    lambda x: {
        'Positive': x.loc[lambda x: x > 0].count() / x.count() * 100,
        'Negative': x.loc[lambda x: x <= 0].count() / x.count() * 100,
    }).unstack()

plt.figure()
plt_table.plot.bar(stacked=True, figsize=(15, 7))
plt.title('Nvidia stock Monthly Upmoves vs. Downmoves by Percentage')
plt.xlabel('Year_month')
plt.ylabel('Percentage of Total Moves')
plt.axhline(50, color='red', linestyle='--', label='50% mark')
plt.legend()

plt.show()
```

```python
#Splitting Data Set into train and test sets
# keeping 2024 data as test , everything before as train(leaving 1st 88 data points as we used 88d returns as a predictor,
# avoiding na in training data)

start = datetime.strftime(df_nvda.index[88], '%Y-%m-%d')
train, test = df_nvda.loc[start:'2023-12-31', :], df_nvda.loc['2024-01-02':, :]
train_x = train.drop(columns=['target', 'return', 'Year_month'])
train_y = train['target']
len(train), len(test)
```

```python
def get_train_test_data(start_dt, end_dt, start_dt_test, end_dt_test):
    """
    Function to iteratively generate train and test sets
    
    """

    train, test = df_nvda.loc[start_dt:end_dt, :], df_nvda.loc[
        start_dt_test:end_dt_test, :]

    train_x = train.drop(columns=['target', 'return', 'Year_month'])
    train_y = train['target']

    test_x = test.drop(columns=['target', 'return', 'Year_month'])
    test_y = test['target']

    return train_x, train_y, test_x, test_y
```

```python
#Setting start and end dates for incrementally fetching train test data

dates_start, dates_end = [], ['2023-12-31']

for i in range(1, 13, 1):
    end_dt = calendar.monthrange(2024, i)[1]
    dt_start = '2024-' + str(i).zfill(2) + str('-') + '01'
    dt_end = '2024-' + str(i).zfill(2) + str('-') + str(end_dt)
    dates_start.append(dt_start)
    dates_end.append(dt_end)
```

```python
#Step 2: Fitting tuned model on train data to fetch predictions on unseen data 
```

```python
# tuned parameters taken from hyper parameter tuning done in the next section

params_rf =  {'ccp_alpha': 0.0001,
              'criterion': 'gini',
              'max_depth': 2,
              'min_samples_leaf': 20,
              'min_samples_split': 60,
              'n_estimators': 10}

start_dt = start


def fit_rf(train_x, train_y, test_x):
    
    """
    Function to iteratively train the model on train data set and return prediction on unseen test data  
    
    """
    
    # creating a RF classifier
    clf = RandomForestClassifier(**params_rf)       
    
    clf.fit(train_x, train_y)
    y_pred = clf.predict(test_x)
    
    return y_pred
```

```python
output_rf = pd.DataFrame()
for i in range(12):
    train_x, train_y, test_x, test_y = get_train_test_data(
        start_dt, dates_end[i], dates_start[i], dates_end[i + 1])
    
    y_pred = fit_rf(train_x, train_y, test_x)
    
    final_output = pd.DataFrame({
        'Date': test_y.index.tolist(),
        'Signal': y_pred
    })
    output_rf = pd.concat([output_rf, final_output], axis=0)
```

```python
def finding_strategy_returns(output_model, test): 
    
    """
    
    
    """
    test['delta_close'] = test['Close'].diff()
    test_returns = test[['Close','ret_1d', 'delta_close']].iloc[1:]
    
    output = output_model.copy().set_index('Date')
    output['Signal'] = np.where(output['Signal']<1, -1, 1) #changing prediction of down move to -1 to facilitate pnl calc.
    shifted_output = output.shift(1).dropna() #shifting the output forward for pnl calc. assuming long/short trade made as per pred

    


    # Returns basis on machine learning model generated trade signals
    returns = pd.concat([shifted_output,test_returns], axis=1)
    returns['pnl'] = returns['Signal']*returns['ret_1d']
    returns['dollar_pnl'] = returns['Signal']*returns['delta_close']
    
    
    return returns
```

```python

test = df_nvda.loc['2024-01-02':, :]
returns =  finding_strategy_returns(output_rf, test)

if len(returns):
    # Cumulative Returns
    cumulative_returns = (returns['pnl'] + 1).cumprod()

    # Sharpe Ratio
    # Assuming risk-free rate as 0 for simplicity
    daily_rf_rate = 0
    annual_rf_rate = daily_rf_rate * 252
    strategy_volatility = returns['pnl'].std() * np.sqrt(252)
    sharpe_ratio = (returns['pnl'].mean() - annual_rf_rate) / strategy_volatility


   # Drawdown
    cum_max = cumulative_returns.cummax()
    drawdown = (cumulative_returns - cum_max) / cum_max
    max_drawdown = drawdown.min()


   # Print the results
    print("Cumulative Returns:")
    print(cumulative_returns[-1] if not cumulative_returns.empty else "No trades executed.")
    print("\nSharpe Ratio:")
    print(sharpe_ratio)
    print("\nMax Drawdown:")
    print(max_drawdown)
else:
    print("No trades executed. Cannot compute performance metrics.")
```

```python
"""
    Title: Bollinger Band Strategy (NSE)
    Description: This is a long short strategy based on bollinger bands
        and SMA dual signals
    Style tags: Systematic Fundamental
    Asset class: Equities, Futures, ETFs and Currencies
    Dataset: NSE
"""
from blueshift.library.technicals.indicators import bollinger_band, ema, rsi

from blueshift.finance import commission, slippage
from blueshift.api import(  symbol,
                            order_target,
                            set_commission,
                            set_slippage,
                            schedule_function,
                            date_rules,
                            time_rules,
                       )

def initialize(context):
    """
        A function to define things to do at the start of the strategy
    """
    # universe selection
    context.securities = [symbol('NIFTY-I'),symbol('BANKNIFTY-I')]

    # define strategy parameters
    context.params = {'indicator_lookback':375,
                      'indicator_freq':'1m',
                      'buy_signal_threshold':0.5,
                      'sell_signal_threshold':-0.5,
                      'SMA_period_short':45,
                      'SMA_period_long':60,
                      'BBands_period':300,
                      'trade_freq':5,
                      'leverage':2}

    # variables to track signals and target portfolio
    context.signals = dict((security,0) for security in context.securities)
    context.target_position = dict((security,0) for security in context.securities)

    # set trading cost and slippage to zero
    set_commission(commission.PerShare(cost=0.0, min_trade_cost=0.0))
    set_slippage(slippage.FixedSlippage(0.00))
    
    freq = int(context.params['trade_freq'])
    schedule_function(run_strategy, date_rules.every_day(),
                      time_rules.every_nth_minute(freq))
    
    schedule_function(stop_trading, date_rules.every_day(),
                      time_rules.market_close(minutes=30))
    
    context.trade = True
    
def before_trading_start(context, data):
    context.trade = True
    
def stop_trading(context, data):
    context.trade = False

def run_strategy(context, data):
    """
        A function to define core strategy steps
    """
    if not context.trade:
        return
    
    generate_signals(context, data)
    generate_target_position(context, data)
    rebalance(context, data)

def rebalance(context,data):
    """
        A function to rebalance - all execution logic goes here
    """
    for security in context.securities:
        qty = security.mult*context.target_position[security]
        order_target(security, qty)

def generate_target_position(context, data):
    """
        A function to define target portfolio
    """
    # trade a single lot
    weight = 1

    for security in context.securities:
        if context.signals[security] > context.params['buy_signal_threshold']:
            context.target_position[security] = weight
        elif context.signals[security] < context.params['sell_signal_threshold']:
            context.target_position[security] = -weight
        else:
            context.target_position[security] = 0


def generate_signals(context, data):
    """
        A function to define define the signal generation
    """
    try:
        price_data = data.history(context.securities, 'close',
            context.params['indicator_lookback'],
            context.params['indicator_freq'])
    except Exception as e:
        raise
        return

    for security in context.securities:
        px = price_data.loc[:,security].values
        context.signals[security] = signal_function(px, context.params)

def signal_function(px, params):
    """
        The main trading logic goes here, called by generate_signals above
    """
    upper, mid, lower = bollinger_band(px,params['BBands_period'])
    if upper - lower == 0:
        return 0
    
    ind2 = ema(px, params['SMA_period_short'])
    ind3 = ema(px, params['SMA_period_long'])
    last_px = px[-1]
    dist_to_upper = 100*(upper - last_px)/(upper - lower)

    if dist_to_upper > 95:
        return -1
    elif dist_to_upper < 5:
        return 1
    elif dist_to_upper > 40 and dist_to_upper < 60 and ind2-ind3 < 0:
        return -1
    elif dist_to_upper > 40 and dist_to_upper < 60 and ind2-ind3 > 0:
        return 1
    else:
        return 0
```

```python
"""
    Title: Bollinger Band Strategy (NSE)
    Description: This is a long short strategy based on bollinger bands
        and RSI based signals
    Style tags: Systematic Fundamental
    Asset class: Equities, Futures, ETFs and Currencies
    Dataset: NSE
"""
from blueshift.library.technicals.indicators import bollinger_band, ema, rsi

from blueshift.finance import commission, slippage
from blueshift.api import(  symbol,
                            order_target,
                            set_commission,
                            set_slippage,
                            schedule_function,
                            date_rules,
                            time_rules,
                       )

def initialize(context):
    """
        A function to define things to do at the start of the strategy
    """
    # universe selection
    context.securities = [symbol('NIFTY-I'),symbol('BANKNIFTY-I')]

    # define strategy parameters
    context.params = {'indicator_lookback':375,
                      'indicator_freq':'1m',
                      'buy_signal_threshold':0.5,
                      'sell_signal_threshold':-0.5,
                      'SMA_period_short':45,
                      'SMA_period_long':60,
                      'BBands_period':300,
                      'trade_freq':5,
                      'leverage':2, 
                      'rsi_period' : 14,  # RSI lookback period
                      'rsi_overbought' : 70,
                      'rsi_oversold' : 30
                      }

    # variables to track signals and target portfolio
    context.signals = dict((security,0) for security in context.securities)
    context.target_position = dict((security,0) for security in context.securities)

    # set trading cost and slippage to zero
    set_commission(commission.PerShare(cost=0.0, min_trade_cost=0.0))
    #set_slippage(slippage.FixedSlippage(0.00))
    set_slippage(slippage.Volumesharemodel
    
    freq = int(context.params['trade_freq'])
    schedule_function(run_strategy, date_rules.every_day(),
                      time_rules.every_nth_minute(freq))
    
    schedule_function(stop_trading, date_rules.every_day(),
                      time_rules.market_close(minutes=30))
    
    context.trade = True
    
def before_trading_start(context, data):
    context.trade = True
    
def stop_trading(context, data):
    context.trade = False

def run_strategy(context, data):
    """
        A function to define core strategy steps
    """
    if not context.trade:
        return
    
    generate_signals(context, data)
    generate_target_position(context, data)
    rebalance(context, data)

def rebalance(context,data):
    """
        A function to rebalance - all execution logic goes here
    """
    for security in context.securities:
        qty = security.mult*context.target_position[security]
        order_target(security, qty)

def generate_target_position(context, data):
    """
        A function to define target portfolio
    """
    # trade a single lot
    weight = 1

    for security in context.securities:
        if context.signals[security] > context.params['buy_signal_threshold']:
            context.target_position[security] = weight
        elif context.signals[security] < context.params['sell_signal_threshold']:
            context.target_position[security] = -weight
        else:
            context.target_position[security] = 0


def generate_signals(context, data):
    """
        A function to define define the signal generation
    """
    try:
        price_data = data.history(context.securities, 'close',
            context.params['indicator_lookback'],
            context.params['indicator_freq'])
    except Exception as e:
        raise
        return

    for security in context.securities:
        px = price_data.loc[:,security].values
        context.signals[security] = signal_function(px, context.params)

def signal_function(px, params):
    """
        The main trading logic goes here, called by generate_signals above
    """
    upper, mid, lower = bollinger_band(px,params['BBands_period'])
    if upper - lower == 0:
        return 0
    
    ind2 = ema(px, params['SMA_period_short'])
    ind3 = ema(px, params['SMA_period_long'])    
    last_px = px[-1]
    dist_to_upper = 100*(upper - last_px)/(upper - lower)
    rsi_value  =  rsi(px, 14)
    
    if dist_to_upper > 95:
        return -1
    elif dist_to_upper < 5:
        return 1
    # elif dist_to_upper > 40 and dist_to_upper < 60 and ind2-ind3 < 0:
    #     return -1
    # elif dist_to_upper > 40 and dist_to_upper < 60 and ind2-ind3 > 0:
    #     return 1
    elif dist_to_upper > 40 and dist_to_upper < 60 and rsi_value > 70:
        return -1
    elif dist_to_upper > 40 and dist_to_upper < 60 and rsi_value < 30:
        return 1
    else:
        return 0
```
